```python
import numpy as np
import pandas as pd
import matplotlib as plt
import array
from itertools import product

data = pd.read_csv('BASKETBALLPLAYERDATA.xlsx.csv', encoding='latin1')

def classify_role(mp):
    if mp >= 26.5:
        i =  'Starter'
    else:
       i = 'Bench Player'
    return i   


data['Role'] = data['MP'].apply(classify_role)

# print(data)
# print(data.Role)  
# print(data[data.Role == 'Starter']) 

def GoldenState(z):
    if z == "GSW":
     a = 1
    else :
      a = 0;  
    return a 

# def filterTeamSTarters(df, team_name):
#     filtered_data = df[(df["Tm"] == team_name) & (df['Role'] == 'Starter')]
#     return filtered_data

#print(filterTeamSTarters(data,"TOR"))
    

data["Warrior"] = data["Tm"].apply(GoldenState)


# def PlayersOn(x):
    
#  if (data[(data['Warrior'] == 1) & (data['Role'] == 'Starter')]):
#     data[data]


def add_on_the_floor_column(df, team_name):
    # Initialize the 'On the Floor' column with zeros
    df['On the Floor'] = 0
    
    # Filter the dataframe for the specified team
    team_df = df[df['Tm'] == team_name].copy()
    
    # Track players by position
    positionsNeeded = {'PG': 1, 'SG': 1, 'SF': 1, 'PF': 1, 'C': 1}
    players_on_floor = []
    
    # Iterate through the team dataframe
    for index, row in team_df.iterrows():
        pos = row['Pos']
        if positionsNeeded.get(pos, 0) > 0 and len(players_on_floor) < 5:
            df.at[index, 'On the Floor'] = 1
            positionsNeeded[pos] -= 1
            players_on_floor.append(row['Player'])
        else:
            df.at[index, 'On the Floor'] = 0
    
    # If more than 5 players are marked as 'On the Floor', reset to 0
    if len(players_on_floor) > 5:
        df.loc[df['Tm'] == team_name, 'On the Floor'] = 0
    print(players_on_floor)
    return df


data['On the Floor'] = 0

# Get unique team names
teams = data['Tm'].unique()

# List to collect updated DataFrames
updated_dataframes = []


# Iterate over each team
for team in teams:
    
    # Filter the DataFrame for the current team
    team_data = data[data['Tm'] == team].copy()
    
    # Apply the function to assign players on the floor for the current team
    updated_team_data = add_on_the_floor_column(team_data, team)
    
    # Append the updated team DataFrame to the list
    updated_dataframes.append(updated_team_data)
    

# Concatenate all updated DataFrames back into a single DataFrame
data = pd.concat(updated_dataframes)

# Reset index to ensure it is unique
data.reset_index(drop=True, inplace=True)


#print(data[(data['On the Floor'] == 1)])

```

```python
from itertools import product

def generate_five_man_combos(df):
    teams = df['Tm'].unique()  # Get unique teams
    result = []  # Initialize the result list

    for team in teams:
        if team == "TOT":
            continue  # Skip the "TOT" team

        team_df = df[df['Tm'] == team]
        
        pgs = team_df[team_df['Pos'] == 'PG']['Player'].tolist()
        sgs = team_df[team_df['Pos'] == 'SG']['Player'].tolist()
        sfs = team_df[team_df['Pos'] == 'SF']['Player'].tolist()
        pfs = team_df[team_df['Pos'] == 'PF']['Player'].tolist()
        cs = team_df[team_df['Pos'] == 'C']['Player'].tolist()
        
        if pgs and sgs and sfs and pfs and cs:
            combos = list(product(pgs, sgs, sfs, pfs, cs))
            for combo in combos:
                result.append({
                    'Team': team,
                    'PG': combo[0],
                    'SG': combo[1],
                    'SF': combo[2],
                    'PF': combo[3],
                    'C': combo[4]
                })

    return pd.DataFrame(result)
combinations = generate_five_man_combos(data)
print(combinations)
    
```

```python
def assign_play(lineup):
    if ((lineup['PG']['AST'] >= 4) and (lineup['PG']['PTS'] <= 22) and lineup['PF']['PTS'] >= 20):
        return "Pick and Roll"
    elif lineup['PG']['PTS'] >= 26 and lineup['SG']['PTS'] <= 20 and lineup['PF']['PTS'] <= 15:
        return "Isolation"
    elif lineup['PG']['PTS'] <= 15 and lineup['SG']['AST'] >= 5 and lineup['C']['PTS'] >= 15:
        return "Motion Offense"
    elif lineup['SG']['PTS'] >= 22 and lineup['SG']['STL'] <= 20 and lineup['PF']['DRB'] >= 1.8:
        return "Flare Screen"
    elif lineup['PG']['STL'] >= 1 and lineup['SF']['DRB'] >= 4 and lineup['C']['BLK'] >= 0.8:
        return "Pop out screen"
    elif lineup['PG']['STL'] >= 1 and lineup['SF']['PTS'] >= 16 and lineup['C']['BLK'] <= 0.7:
        return "Screen and Cut"
    elif lineup['PF']['STL'] >= 0.75 and lineup['SG']['DRB'] <= 5 and lineup['PF']['BLK'] >= 0.68:
        return "Kick out midrange"
    elif lineup['PG']['STL'] >= 1 and lineup['SF']['DRB'] >= 4 and lineup['C']['BLK'] >= 0.8:
        return "Back Screen Post"
    elif lineup['SG']['PTS'] >= 23 and lineup['PG']['AST'] >= 7 and lineup['C']['ORB'] >= 1.7:
        return "Piston Elevator"
    else:
        return "Floppy"
```

```python
def merge_player_data(row, player_data):
    lineup = {}
    for pos in ['PG', 'SG', 'SF', 'PF', 'C']:
        player_name = row[pos]
        player_stats = player_data[player_data['Player'] == player_name].iloc[0].to_dict()
        lineup[pos] = player_stats
    return lineup
```

```python
combinations['Play'] = combinations.apply(lambda row: assign_play(merge_player_data(row, data)), axis=1)
print(combinations)
```

```python
def assignGSplay(lineup):
    if lineup['PG']['MP'] >= 28.5  and lineup['PG']['FG%'] >= 0.41 and lineup['PF']['FG'] >= 5:
        return "GS Pick and Roll"
    elif lineup['PG']['PTS'] >= 26 and lineup['SG']['PTS'] <= 19 and lineup['PF']['PTS'] <= 12:
        return "GS Isolation"
    elif lineup['PG']['PTS'] <= 15 and lineup['SG']['TOV'] >=2.4 and lineup['C']['PTS'] >= 15:
        return "GS Motion Offense"
    elif lineup['SG']['PTS'] >= 22 and lineup['PG']['AST'] >=5 and lineup['PF']['ORB'] >= 1.8:
        return "GS ELevator Screen"
    elif lineup['SG']['3P%'] >= 0.375 or lineup['PG']['3P%'] >= 0.405 and lineup['PG']['AST'] >= 3 and lineup['PF']['ORB'] >= 1.8:
        return "GS Floppy"     
    else:
        return "GS Triangle Offense"
```

```python
def meVSthem(lineup):
    if lineup["PG"]["Tm"] == "GSW":
        return assignGSplay(lineup)
    elif lineup["PG"]["Tm"] == "TOT":
        return "Invalid Team"
    else:
        return assign_play(lineup)
```
